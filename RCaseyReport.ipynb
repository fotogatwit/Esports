```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
import matplotlib.pyplot as plt
esports = pd.read_csv('Esports.csv')
```

Many competitve tournaments take place around the world, causing players to move, or travel all over the world to compete. However, many tournaments take place in relatively the same places, as shown below in the graph, most tournaments take place in places in the US and China, or most recently, all online tournaments. This has caused some small issues in the past, what with some tournaments requiring people to move to where the tournament is, which people haven't been happy about. Many of the League of Legends and CS:GO tournaments take place in the US, which is mainly why the graph favors the US so heavily in this dataset.

```python
with sns.axes_style('white'):
    g = sns.catplot( data=esports,x="Country" , kind = "count").set(title = "Tournaments per game")
    g.set_xticklabels(rotation = 90)
```
