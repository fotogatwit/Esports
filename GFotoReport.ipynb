```python
import numpy as np
import pandas as pd
import seaborn as sns
sns.set(rc={'figure.figsize':(11.7,8.27)})
import matplotlib.pyplot as plt
esports = pd.read_csv('Esports.csv')
```

Hello, First of all I would like to start of with the visual depicting when the esports tournaments took place.
As we can see, we have a small rise from 2005 till about 2015. Esports wasn't that big around then, so thier prize pools werent enough to crack the top 200. But once we hit 2016, we can see a large rise until we reach about 2020. So around 2016, esports was starting to become a big thing so we can obisouly see that it finally got enough backing to give out a large sum of money. We see the graph start to fall down again in 2020 and 2021. This is obisouly a reflection of covid 19 hitting, as tournaments were canceled or even hosted online which would have dropped the prize pool significatly since they would be losing out on ticket sales. Overall, this graph just shows us exactly when esports started to gain a large backing and large amount of funds that they were able to give away.

```python
with sns.axes_style('white'):
    g = sns.catplot( data=esports,x="Year" , kind = "count").set(title = "Tournments Every Year")
    g.set_xticklabels(rotation = 90)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(esports['Year'][:, np.newaxis], esports["TotalUSDPrize"])

#predict on non-noisy data:
xfit = esports['Year']
yfit = model.predict(xfit[:, np.newaxis])

plt.title("Linear Regression for Prizepool of Every Year")
plt.xlabel('Year') 
plt.ylabel('TotalUSDPrize')
plt.text(2005, 33000000, 'Slope:', fontsize = 20, bbox = dict(facecolor = 'blue', alpha = 0.5))
plt.text(2005, 30000000, 'Y = -94786.29X + 193353665.12', fontsize = 20, bbox = dict(facecolor = 'blue', alpha = 0.5))

plt.scatter(esports['Year'], esports["TotalUSDPrize"]);
plt.plot(xfit, yfit);
```

```python
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)

x = model.coef_[0]
b = model.intercept_
```
